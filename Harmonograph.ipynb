# Harmonograph Implementation Using Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameter that will be used
n = 1000000
t = np.logspace(np.log10(10), np.log10(500), n)

#change this parameter
A = [3, 6, 1.5, 1.5]
d = [.004, .001, .002, .0015]
f = [3, 1, 2, 2.5]

#create X & Y
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)

#show plot
plt.plot(x,y,'g',linewidth=.1)
plt.axis('off')
plt.show()
```

```python
#see the component one by one
plt.plot(np.sin(t*f[0])*np.exp(-d[0]*3*t),linewidth=.5)
plt.show()
```

```python
fig= plt.figure(figsize=(3,6))
axes= fig.add_axes([5,5,4,0.9])
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)
axes.plot(x,y,y,x,'r',linewidth=.4)
plt.show()
```
